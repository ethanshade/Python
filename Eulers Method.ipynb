```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def SIR(S0, I0, R0, beta, gamma, t_max, stepsize):
    T = np.arange(0,t_max+stepsize,stepsize)
    S = np.zeros(len(T))
    I = np.zeros(len(T))
    R = np.zeros(len(T))
    N = S0+I0+R0

    for idx,t in enumerate(T):
        if idx==0:
            S[idx] = S0
            I[idx] = I0
            R[idx] = R0
        else:
            dS_dt = -beta*S[idx-1]*I[idx-1]/N
            dI_dt = beta*S[idx-1]*I[idx-1]/N - gamma*I[idx-1]
            dR_dt = gamma*I[idx-1]

            S[idx] = S[idx-1]+dS_dt*stepsize
            I[idx] = I[idx-1]+dI_dt*stepsize
            R[idx] = R[idx-1]+dR_dt*stepsize

    return S,I,R,T
```

```python
# def SIR(S0, I0, R0, beta, gamma, t_max, stepsize):
S,I,R,T = SIR(500,1,0,
              1,.1,50,0.05)
fig, ax = plt.subplots()
ax.plot(T,S, color="b", label = "Suseptible")
ax.plot(T,I, color="r", label = "Infected")
ax.plot(T,R, color="k", label = "Recovered")
ax.set_xlabel("Time")
ax.set_ylabel("People")
ax.legend()
ax.set_xlim(left=0)
ax.set_ylim(bottom=0)
```
